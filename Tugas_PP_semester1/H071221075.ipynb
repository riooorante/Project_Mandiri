```python
import pandas as pd

# Membaca file BPKSD.csv
data = pd.read_csv("BPKSD.csv")
data
```

```python
# Menghilangkan kolom pertama (Timestamp)
data = data.iloc[:,1:] 
data
```

```python
# Konversi Data
Konversi = {"SS":4, "S":3, "TS":2, "STS":1}
data = data.replace(Konversi)
data
```

```python
# Menampilkan Data Opini atau Kenyataan
DataOpini = data.iloc[:,[2*i for i in range(0,21)]]
DataOpini
```

```python
# Menampilkan data Harapan
DataHarapan = data.iloc[:,[2*i+1 for i in range(0,21)]]
DataHarapan
```

```python
import numpy as np

# Mencari data rata-rata setiap pertanyaan (Kenyataan/Opini)
rataanOpiniP = DataOpini.mean(axis=0)

# Mencari data rata-rata setiap pertanyaan (Harapan)
rataanHarapanP = DataHarapan.mean(axis=0)

# Mencari Bobot dan W x X
WPertanyaan = (np.array(rataanHarapanP) / np.array(rataanHarapanP).sum())*100
WX_Pertanyaan = [x*y for x,y in zip(WPertanyaan, rataanOpiniP)]

# Membuat Tabel
KontenPertanyaan = {r'Kenyataan (X)':  [x for x in rataanOpiniP],
        r'Harapan (Y)': [x for x in rataanHarapanP], r"Bobot (W)":[x for x in WPertanyaan],r"W x X":[x for x in WX_Pertanyaan]}

tabelPertanyaan = pd.DataFrame(KontenPertanyaan, index=[x+1 for x in range(21)])
tabelPertanyaan
```

```python
CSI_Pertanyaan = sum(WX_Pertanyaan) / 4
if CSI_Pertanyaan >= 80.01 and CSI_Pertanyaan <= 100.00:
    X = "Sangat Puas"
elif CSI_Pertanyaan >= 65.01 and CSI_Pertanyaan <= 80.00:
    X = "Puas"
elif CSI_Pertanyaan >= 50.01 and CSI_Pertanyaan <= 65.00:
    X = "Cukup Puas"
elif CSI_Pertanyaan >= 35.01 and CSI_Pertanyaan <= 50.00:
    X = "Kurang Puas"
else:
    X = "Tidak Puas"
print(f"CSI = {X}")
```

```python
# Menampilkan grafik 
from matplotlib import pyplot as plt
plt.title("Indicator Performance Analysis (IPA) Pertanyaan")
plt.xlabel("Kenyataan (X)")
plt.ylabel("Harapan (Y)")
plt.scatter(rataanOpiniP, rataanHarapanP)

# Garis Horizontal
plt.plot([rataanOpiniP.min()-0.1,rataanOpiniP.max()+0.1],[rataanHarapanP.mean(),rataanHarapanP.mean()], color='red')

# Garis Vertikal
plt.plot([rataanOpiniP.mean(),rataanOpiniP.mean()],[rataanHarapanP.min()-0.1,rataanHarapanP.max()+0.1], color='black')

plt.grid()

# Menambahkan legenda Grafik
plt.legend(["Data" , "Harapan", "Kenyataan"], loc = "lower left")

# Menampilkan Grafik
plt.show()
```
